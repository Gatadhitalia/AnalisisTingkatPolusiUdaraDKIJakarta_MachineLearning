# 2. Data Cleaning
## a. Data Cleaning dengan Penghapusan Baris
Karena terdapat banyak data yang null, maka pada langkah ini kita akan menghapus baris yang memiliki nilai null.

```python
data1 = data1.dropna(subset=['pm25'], inplace=False)
data2 = data2.dropna(subset=['pm25'], inplace=False)
data3 = data3.dropna(subset=['pm25'], inplace=False)
data4 = data4.dropna(subset=['pm25'], inplace=False)
data5 = data5.dropna(subset=['pm25'], inplace=False)
```

## a. Data Cleaning dengan Pengisian nilai
Pada langkah ini kita akan mengisi baris yang memiliki nilai null.
Data numerik yang null akan di isi dengan nilai median dari data tersebut.

```python
data2['pm10']=data2['pm10'].fillna((data2['pm10'].median()))
data2['so2']=data2['so2'].fillna((data2['so2'].median()))
data2['co']=data2['co'].fillna((data2['co'].median()))
data2['o3']=data2['o3'].fillna((data2['o3'].median()))
data2['no2']=data2['no2'].fillna((data2['no2'].median()))
```
